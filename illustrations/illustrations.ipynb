```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function to be optimized
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
```

```python
# Create a grid of values
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
x, y = np.meshgrid(x, y)
z = f(x, y)

# Plot the surface and contours
fig = plt.figure(figsize=(12, 6))

# First subplot for the 3D surface plot
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(x, y, z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
# ax1.set_title('Surface plot')

# Second subplot for the contour plot
# ax2 = fig.add_subplot(122)
# contour = ax2.contourf(x, y, z, levels=50, cmap='viridis')
# fig.colorbar(contour, ax=ax2, shrink=0.5, aspect=5)
# ax2.set_title('Contour plot')

# Show the plots
plt.show()
```
